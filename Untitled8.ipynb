```python
#que1.1

from requests_html import HTMLSession
from bs4 import BeautifulSoup


ht = HTMLSession()
url = "https://en.wikipedia.org/wiki/Egypty"

def data(url):
  gt = ht.get(url)
  soup = BeautifulSoup(gt.text,"html.parser")
  return soup
soup = data(url)
print(soup)
```

```python
#que1.2
def title(soup):
  titles = soup.select(".mw-page-title-main")
  return titles

print(title(soup))
```

```python
#qu1.3
def paragraphs():
  for head in soup.find_all("p"):
    print(head.get_text())
paragraphs()   
```

```python
#que 1.4
def links():
  links = {}
  for link in soup.find_all("a"):
    url = link.get("href",  "")
    if "/wiki/" in url:
      link[link.text.strip()] = url
    print(links, "\n")  

links()
```

```python
#que1.5
from requests_html import HTMLSession
from bs4 import BeautifulSoup

def scraper(wiki):

  def data(url):
    gt = ht.get(url)
    soup = BeautifulSoup(gt.text, 'html.parser')
    return soup

  soup = data(wiki)

  def title():
    titles = soup.find("title")
    print(titles,"\n")

  def paragraphs():
   for head in soup.find_all("p"):
     print(head.get_text(),"\n")
  
  def links():
    links = {}
    for link in soup.find_all("a"):
      url = link.get("href", "")
      if "/wiki/" in url:
        links[link.text.strip()] = url
    print(links,"\n")
  

  

  print(title(),paragraphs(),links())

scraper("https://en.wikipedia.org/wiki/Egypt")
```
